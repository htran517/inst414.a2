```python
%matplotlib inline

import networkx as nx
import json

g = nx.Graph()
```

```python
with open('newspapers.json', 'r') as file:
    for line in file:
        newspapers = json.loads(line)

newspapers
```

```python
for paper in newspapers['newspapers']:
    g.add_node(paper['title'])

print('Nodes:', len(g.nodes))

counter = 0

for left_paper in newspapers['newspapers']:
        for right_paper in newspapers['newspapers'][counter+1:]:
                if left_paper['state'] == right_paper['state']:
                    if g.has_edge(left_paper['title'], right_paper['title']):
                        continue
                    g.add_edge(left_paper['title'], right_paper['title'])      
        counter += 1
        
nx.ego_graph(g, n = 'Birmingham state herald.', radius = 3)

nx.write_graphml(g, 'news.graphml')
```
